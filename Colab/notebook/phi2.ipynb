```python
from transformers import AutoTokenizer, AutoModelForCausalLM

```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-1.5", trust_remote_code=True)
```

```python
pip install einops
```

```python
model = AutoModelForCausalLM.from_pretrained("microsoft/phi-1_5", trust_remote_code=True)
```

```python
# Encode some text to generate from.
input_ids = tokenizer.encode("""<s>[INST]<<SYS>>
Give me associations between metrics and values in the given text.
Give me word sense map.
Don't give explanation.
Generate a JSON.
<</SYS>>
Input string = "top 10 accounts"[/INST]
""", return_tensors='pt')

# Generate text using Phi-1.5 model
output = model.generate(input_ids, max_length=100)

# Decode the generated text
decoded_output = tokenizer.decode(output[0], skip_special_tokens=True)
print(decoded_output)
```
